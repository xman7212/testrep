<h3>Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
import interp as lag
import cubic_spline_demo as cube

def lagrangeEval(f,N,a,b):
    xint = np.linspace(a,b,N+1)
    yint = f(xint)

    Neval = 1000
    xeval = np.linspace(a,b,Neval+1)
    yeval_l= np.zeros(Neval+1)

    for i in range(Neval+1):
        yeval_l[i]= lag.eval_lagrange(xeval[i],xint,yint,N)
    y = f(xeval)
    plt.plot(xeval,y, label = "Exact")
    plt.plot(xeval,yeval_l, label = "evaluated at "+str(N))
    plt.ylim(-1.5,1.5)
    plt.title("n="+str(N))
    plt.legend()
    plt.show()
    plt.close()
    
    
```

<h4>Lagrange:

```python
f = lambda x: 1/(1+x**2)
N=[5,10,15,20]
a=-5
b=5
for n in N:

    lagrangeEval(f,n,a,b)
```

<h4>Hermite:

```python
def hermite(f,N,a,b):
    xint = np.linspace(a,b,N+1)
    yint = f(xint)

    Neval = 1000
    xeval = np.linspace(a,b,Neval+1)
    yeval= np.zeros(Neval+1)

    e,h,g = cube.create_natural_spline(yint,xint,N)
    yeval = cube.eval_cubic_spline(xeval,Neval,xint,N,e,h,g)
    fex = f(xeval)
    nerr = np.linalg.norm(fex-yeval)
    print('nerr = ', nerr)
    plt.figure()
    plt.plot(xeval,fex,'ro-',label='exact function')
    plt.plot(xeval,yeval,'bs--',label='natural spline')
    plt.legend
    plt.show()
    err = abs(yeval-fex)
    plt.figure()
    plt.semilogy(xeval,err,'ro--',label='absolute error')
    plt.legend()
    plt.show()
```

```python
hermite(f,5,a,b)
```
