```python
import numpy as np
import monomial_interp as mi
import matplotlib.pyplot as plt
import interp as lag
```

```python
def interp_for_lab(N,plot):
    f31 = lambda x:1/(1+(10*x)**2)

    a=-1
    b=1
    j=np.linspace(0,N+1,N+1)
    x31 = -1+(j-1)*(2/(N-1))
    y31=f31(x31)
    V = mi.Vandermonde(x31,N)
    Vinv = mi.inv(V)
    coef = Vinv @ y31
    Neval = 100    
    xeval = np.linspace(a,b,Neval+1)
    yeval = mi.eval_monomial(xeval,coef,N,Neval)
    y311 = f31(xeval)
    if plot == "actual":
        plt.plot(xeval,yeval, label = "N="+str(N))
        plt.title("Monomial Interpolation up to N ="+str(N))
    elif plot =="abs":
        plt.plot(xeval,yeval-y311, label = "N="+str(N))
        plt.title("Monomial Interpolation Absolute Error up to N ="+str(N))
    elif plot =="rel":
        plt.plot(xeval,(yeval-y311)/y311, label = "N="+str(N))
        plt.title("Monomial Interpolation Relative Error up to N ="+str(N))
    
    plt.legend()
    return

def lag_for_lab(N,plot):
    f31 = lambda x:1/(1+(10*x)**2)

    a=-1
    b=1
    j=np.linspace(0,N+1,N+1)
    x31 = -1+(j-1)*(2/(N-1))
    y31=f31(x31)
    V = mi.Vandermonde(x31,N)
    Vinv = mi.inv(V)
    coef = Vinv @ y31
    Neval = 1000    
    xeval = np.linspace(a,b,Neval+1)
    yeval= np.zeros(Neval+1)
    

    for kk in range(Neval+1):
       yeval[kk] = lag.eval_lagrange(xeval[kk],x31,y31,N)
       

    y311 = f31(xeval)
    if plot == "actual":
        plt.plot(xeval,yeval, label = "N="+str(N))
        plt.title("Monomial Interpolation up to N ="+str(N))
    elif plot =="abs":
        plt.plot(xeval,yeval-y311, label = "N="+str(N))
        plt.title("Monomial Interpolation Absolute Error up to N ="+str(N))
    elif plot =="rel":
        plt.plot(xeval,(yeval-y311)/y311, label = "N="+str(N))
        plt.title("Monomial Interpolation Relative Error up to N ="+str(N))
    
    plt.legend()
    return
```

```python
N=2
while N<11:
    interp_for_lab(N,"abs")
    N+=1
```

```python
N=2
while N<11:
    interp_for_lab(N,"rel")
    N+=1
```

```python
for N in range(2,11):
    lag_for_lab(N,"abs")
```

```python
for N in range(2,11):
    lag_for_lab(N,"rel")
```
